```python
# Initialize or Load tree
def init_tree(system_name, timing_report):
    tree = Tree(system_name)
    tree.parse_timing_report(timing_report)
    return tree

def pickle_load_tree(file_name):
    with open(file_name, 'rb') as file:
        tree = pickle.load(file)
        return tree
```

```python
class Module:
    def __init__(self, inst_name):
        self.inst_name = inst_name
        self.module_name = ''
        self.count = 0
        self.isModule = True
        self.isLastLevelModule = False
        self.slack = 0
        self.children = []
        
    def add_child(self, child):
        self.children.append(child)
        self.count += 1
    
    def node_slack(self, slack):
        self.slack = slack
    
    def got_you(self):
        self.count += 1


class Tree:
    def __init__(self, root_name):
        self.root = Module(root_name)
        self.root.module_name = root_name
        self.inst_module_name = {}
        self.last_level_module = {}
        self.last_level_inst = {}
        
    def parse_timing_report(self, timing_report):
        for line in timing_report:
            path = line.split('/')
            prefix = re.search(r'^[^\s]*', path[-1]).group(0)
            path[-2] = path[-2] + '_' + prefix
            path[-1] = re.findall(r'\d+\.\d+', path[-1])
            #print(path)
            self.add_node(path)  
    
    def add_node(self, path):
        node = self.root
        for name in path[:-1]:
            child = None
            for n in node.children:
                if n.inst_name == name:
                    node.got_you()
                    child = n
                    break
            if child is None:
                child = Module(name)
                node.add_child(child)
            node = child
            if name == path[-3]:
                # print("find last module: " + name)
                node.isLastLevelModule = True
        node.node_slack(path[-1])
        node.isModule = False

    def instance_search(self,module_name,root=None):
        if root is None:
            root = self.root
        if(root.inst_name == module_name):
            print(root.inst_name + ":" + str(root.count))
            return root
        for child in root.children:     
            self.instance_search(module_name, child)
    
    def module_search(self,module_name,root=None):
        if root is None:
            root = self.root
        if(root.module_name == module_name):
            return root.count
        for child in root.children:     
            count = self.module_search(module_name, child)
            if count is not None:
                return count
    
    def get_module_name(self, node=None):
        if node is None:
            node = self.root
        for index, text in enumerate(netlist):
            pattern = r'\b{0}\b'.format(node.module_name)
            if re.search(pattern, text):
                module_start = index
                module_end = netlist.index('endmodule\n', module_start)
                for node_child in node.children:
                    print(f"searching {node.module_name} - {node_child.inst_name}:")
                    if(node_child.isModule == True):
                        for i in range(module_start, module_end):
                            if node_child.inst_name in netlist[i]:
                                node_child_inst_index = netlist[i].find(node_child.inst_name)
                                node_child.module_name = netlist[i][:node_child_inst_index].split()[0]
                        self.get_module_name(node_child)
                    else:
                        break
    
    # store module name from "inst_module.txt"
    def store_module_name(self, node=None):
        if node is None:
            node = self.root
        if node.inst_name in self.inst_module_name:
            node.module_name = self.inst_module_name[node.inst_name]
        for child in node.children:
            self.store_module_name(child)
    
    # save inst_name : module_name to "inst_module.txt"
    def save_inst_module_name(self, node=None):
        if node is None:
            node = self.root
        if(node.isModule == True):
            self.inst_module_name[node.inst_name] = node.module_name   
        else:
            pass
        for node_child in node.children:
            self.save_inst_module_name(node_child)    
    
    # Store last level module
    def store_last_level_module(self, node=None, father_name=''):
        if node is None:
            node = self.root
            father_name = 'system'
        if(node.isLastLevelModule != False):
            last_level_module_name = father_name + '/' + node.module_name
            self.last_level_module[last_level_module_name] = node.count
        else:
            if 'e203' in node.module_name:
                father_name = node.module_name
        for child in node.children:
            self.store_last_level_module(child, father_name)
    
    
    def store_last_level_inst(self, node=None, father_name=''):
        if node is None:
            node = self.root
            father_name = 'system'
        if(node.isLastLevelModule != False):
            last_level_inst_name = father_name + '/' + node.inst_name
            self.last_level_inst[last_level_inst_name] = node.count
        else:
            if 'u_e203' in node.inst_name:
                father_name = node.inst_name
        for child in node.children:
            self.store_last_level_module(child, father_name)
    
    
    # Print tree
    def print_inst_tree(self, node=None, level=0):
        if node is None:
            node = self.root
        if(node.slack != 0):
            print('  '*level + '- ' + node.inst_name + ' -' + str(node.slack[0]))
        else:
            print('  '*level + '- ' + node.inst_name + ' - ' + str(node.count))
        for child in node.children:
            self.print_inst_tree(child, level+1)
    
    # Save tree to "inst_tree.md'
    def save_inst_tree(self, node=None, level=0, max_level=10, file_path=None):
        if node is None:
            node = self.root
        output_str = ''
        if level <= max_level:
            if(node.slack != 0):
                # output_str = '  '*level + '- ' + node.inst_name + ' -' + str(node.slack[0]) + '\n'
                pass
            else:
                output_str = '  '*level + '- ' + node.inst_name + ' - ' + str(node.count) + '\n'
        for child in node.children:
            output_str += self.save_inst_tree(child, level+1, max_level, file_path)
        if file_path is not None:
            with open(file_path, 'w') as f:
                f.write(output_str)
        else:
            print(output_str)
        return output_str
    
    def print_module_tree(self, node=None, level=0):
        if node is None:
            node = self.root
        if(node.slack != 0):
            print('  '*level + '- ' + node.inst_name + ' -' + str(node.slack[0]))
        else:
            print('  '*level + '- ' + node.module_name + ' - ' + str(node.count))
        for child in node.children:
            self.print_module_tree(child, level+1)
    
    # Draw a picture to see the distribution of path
    def draw_path_distribution_vertical(self, filename):
        fig, ax = plt.subplots(figsize=(24, 18))
        x_grouped = []
        y_grouped = []
        last_level_module_name_list = list(self.last_level_module.keys())
        last_level_module_cnt_list = list(self.last_level_module.values())
        for i in range(len(last_level_module_name_list)-1):
            x_grouped.append(last_level_module_name_list[i])
            y_grouped.append(last_level_module_cnt_list[i])
            father_name = last_level_module_name_list[i].split('/')[0]
            next_father_name = last_level_module_name_list[i+1].split('/')[0]
            if father_name != next_father_name:
                x_grouped.append(father_name + '-' + str(self.module_search(father_name)) + '=========================')
                y_grouped.append(0)
        x_grouped.append(last_level_module_name_list[-1])
        y_grouped.append(last_level_module_cnt_list[-1])
        last_name = last_level_module_name_list[-1].split('/')[0]
        x_grouped.append(last_name + '-' + str(self.module_search(last_name)) + '=========================')
        y_grouped.append(0)
        # print(x_grouped)
        # print(y_grouped)
        # print(len(x_grouped) == len(y_grouped))
        ax.bar(x_grouped, y_grouped)
        # ax.bar(list(last_level_module.keys()), list(last_level_module.values()))
        ax.set_title('Critical Paths Distribution', fontdict={'fontsize': 20})
        ax.set_ylabel('Number of Paths', fontsize=20)
        ax.set_xlabel('Module', fontsize=20)
        # ax.set_xticks(list(last_level_module.keys()))
        # ax.set_xticklabels(list(last_level_module.keys()), rotation=45, ha='right')
        ax.set_xticks(range(len(x_grouped)))
        ax.set_xticklabels(x_grouped, rotation=45, ha='right')
        #ax.set_ylim([0, max(list(last_level_module.values())) * 1.2])
        plt.savefig(filename, bbox_inches = 'tight')
        plt.show()
    
        
    # Dump & Load the tree
    def json_dump_inst_module_name(self, file_name):
        with open(file_name, 'w') as file:
            file.write(json.dumps(self.inst_module_name))

    def json_load_inst_module_name(self, file_name):
        with open(file_name, 'r') as file:
            self.inst_module_name = json.load(file)
    
    def json_dump_last_level_module(self, file_name):
        with open(file_name, 'w') as file:
            file.write(json.dumps(self.last_level_module))
    
    def pickle_dump_tree(self, file_name):
        with open(file_name, 'wb') as file:
            pickle.dump(self, file)
```
